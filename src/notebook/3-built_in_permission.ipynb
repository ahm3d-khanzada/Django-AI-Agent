```python
import django_jupyter
django_jupyter.init()
```

```python
from django.contrib.auth import get_user_model
User = get_user_model()

User.objects.all()
```

```python
super_u = User.objects.filter(is_superuser=True).first()
super_u
```

```python
super_u.user_permissions.all()
```

```python
staff_u = User.objects.filter(is_superuser=False , is_staff=True).first()
staff_u
```

```python
staff_u.user_permissions.all()
```

```python
from django.contrib.auth.models import Permission

perms_qs = Permission.objects.filter(codename__icontains="VIEW")
for obj in perms_qs:
    print(obj.codename)
    staff_u.user_permissions.remove(obj)
```

```python
staff_u.user_permissions.set(perms_qs)
```

```python
doc_qs = Permission.objects.filter(codename__endswith="directory")
doc_qs
```

```python
staff_u.user_permissions.set(doc_qs)
```

```python
new_perms = doc_qs | perms_qs
new_perms
```

```python
new_perms = new_perms.exclude(codename__endswith="delete")
new_perms
```

```python
staff_u.user_permissions.set(new_perms)
```
