```python
from permit import PermitApiError

async def get_or_create_resources(resource_data: dict):
    resource_key = resource_data.get("key")

    if not resource_key:
        raise ValueError("resource_data must contain a 'key'")
    try:
        resource = await permit.api.resources.get(resource_key)
    except PermitApiError as e:
        if e.status_code == 404:
            print(f"Resource with key '{resource_key}' not found. Creating new resource.")
            resource = await permit.api.resources.create(resource_data)
        else:
            # Re-raise unexpected errors
            raise

    return resource
```

```python
for resource_data in resourses:
    response = await get_or_create_resources(resource_data)
    print(response)
```

```python
from permit import PermitApiError

async def get_or_create_role(role_data: dict):
    role_key = role_data.get("key")

    if not role_key:
        raise ValueError("role_data must contain a 'key'")

    try:
        role = await permit.api.roles.get(role_key)

    except PermitApiError as e:
        if e.status_code == 404:
            print(f"Role '{role_key}' not found. Creating new role.")
            role = await permit.api.roles.create(role_data)
        else:
            raise

    return role
```

```python
from ai.supervisor import get_supervisor
from langgraph.checkpoint.memory import InMemorySaver
import uuid
```

```python
checkpointer = InMemorySaver()
```

```python
supervisor = get_supervisor(
    model="gpt-4o-mini",
    checkpointer=checkpointer
)
```

```python
doc_id = 24
user_id = user.id
```

```python
config = {
    "configurable": {
        "user_id": user_id,
        "thread_id": str(uuid.uuid4())
    }
}
```

```python
response = supervisor.invoke(
    {
        "messages": [
            {
                "role": "user",
                "content": "Get a list of all documents",
            }
        ]
    },
    config=config,
)
```

```python
from langchain_core.messages import AIMessage

final_answer = None

for msg in response["messages"]:
    if isinstance(msg, AIMessage) and msg.content:
        final_answer = msg.content

print(final_answer)
```
